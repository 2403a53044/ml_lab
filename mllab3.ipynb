```python
#mean squad error
def compute_cost(y,y_pred):
  m=len(y)
  cost=(1/2*m)*np.sum((y-y_pred)**2)
  return cost
```

```python
learning_rate=0.01
epochs=1000
print("learning rate",learning_rate)
print("epochs",epochs)
```

```python
n=len(y)
```

```python
cost_history=[]
```

```python
for epoch in range(epochs):
  y_pred=predict(normalization,a,b)
  da=(1/n)*np.sum((y_pred-y)*normalization)
  db=(1/n)*np.sum(y_pred-y)
  a=a-learning_rate*da
  b=b-learning_rate*db
  cost=compute_cost(y,y_pred)
  cost_history.append(cost)
  if epoch%100==0:
    print("Epoch=",epoch,"cost=",cost)
```

```python
final_slope=a
final_intercept=b
print("Final slope:",final_slope)
print("Final intercept:",final_intercept)
```

```python
def predict_salary(YearsExperience,a,b,x_min_train,x_max_train):
  x_to_predict=np.array(YearsExperience).reshape(-1,1)
  normalization=(x_to_predict-x_min_train)/(x_max_train-x_min_train)
  salary_pred=a*normalization+b
  return salary_pred
```

```python
year=5
predicted_salary=predict_salary(year,final_slope,final_intercept,min,max)
print("Predicted salary for",year,"years of experience:",predicted_salary)
print(predicted_salary[0][0])
```
