```python
import numpy as np
exam_score_array=df['Exam_Score'].to_numpy()
exam_score_array
```

```python
exam_score_array=df['Exam_Score'].to_numpy().reshape(-1,1)
exam_score_array
```

```python
mean=np.mean(exam_score_array)
mean
```

```python
median=np.median(exam_score_array)
median
```

```python
max=np.max(exam_score_array)
max
```

```python
std=np.std(exam_score_array)
std
```

```python
import matplotlib.pyplot as plt
plt.hist(exam_score_array)
plt.show()
```

```python
plt.hist(exam_score_array,bins=100,color='Black')
plt.xlabel('Exam Score')
plt.ylabel('number of students')
plt.title('Distribution  of Exam Scores')
plt.show()
```

```python
scores,count=np.unique(exam_score_array,return_counts=True)
plt.bar(scores,count,color='Green')
plt.xlabel('Exam Score')
plt.ylabel('number of students')
plt.title('Bar chart  of Exam Scores')
plt.show()
```

```python
scores,count=np.unique(exam_score_array,return_counts=True)
plt.pie(count,labels=scores,autopct='%1.1f%%')
plt.title('Pie Chart  of Exam Scores')
plt.show()
```

```python
hourly_studied_array=df['Hours_Studied'].to_numpy()
hourly_studied_array
plt.scatter(hourly_studied_array,exam_score_array)
plt.xlabel('Hours_Studied')
plt.ylabel('Exam_Score')
plt.title('Scatter Plot of Hours_Studied vs Exam_Score')
```
