```python
import numpy as np
```

```python
age_array=df['Age'].to_numpy().reshape(-3,3)
age_array
```

```python
mean_age=np.mean(age_array)
median_age=np.median(age_array)
std_age=np.std(age_array)
print(mean_age,median_age,std_age)
```

```python
cholestrol_array=df['Cholesterol'].to_numpy()
cholestrol_array
```

```python
min_cholestrol=np.min(cholestrol_array)
max_cholestrol=np.max(cholestrol_array)
print(min_cholestrol,max_cholestrol)
```

```python
count_heart_disease=df['Heart Disease'].value_counts()
count_heart_disease
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(age_array,bins=10)
plt.xlabel('Age')
plt.ylabel('no of patients')
plt.title('Histogram of Age')
plt.show()
```

```python
values,counts=np.unique(cholestrol_array,return_counts=True)
plt.bar(values,counts,color='Green')
plt.xlabel('cholestrol level')
plt.ylabel('no of patients')
plt.title('barchart of cholestrol level')
plt.show()
```

```python
plt.scatter(df['Age'],df['Max HR'])
plt.xlabel('Age')
plt.ylabel('max heart rate')
plt.title('Age vs Max HR')
plt.show()
```
